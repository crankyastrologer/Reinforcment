```python
from GridWorld import GridWorld
from PolicyIteration import PolicyIteration
from GridWorld import GridWorld
from PolicyIteration import PolicyIteration

import numpy as np

problem = GridWorld('world02.csv', reward={0: -0.04, 1: 10.0, 2: -2.5, 3: np.NaN}, random_rate=0.2)
problem.plot_map(fig_size=(15, 8))
solver = PolicyIteration(problem.reward_function, problem.transition_model, gamma=0.9, init_policy=None)
solver.train()
problem.plot_policy(policy=solver.policy, fig_size=(10, 8))
problem.random_start_policy(policy=solver.policy, start_pos=(5, 3), n=1000)
```

```python
grid = [['R','R','R',1],
        ['U',0,'U',-1],
        ['U','R','U','L']]
gridon = [[0,0,0,1],
          [0,0,0,-1],
          [0,0,0,0]]
gamma = 0.9

for i in range(0,100):
    diff = 0
    for i in range(0,3):
        for j in range(0,4):
            if grid[i][j]=='R':
                old =  gridon[i][j] - gamma*gridon[i][j+1]
                gridon[i][j] = gamma*gridon[i][j+1]
                diff +=old

            if grid[i][j]=='L':
                old = gridon[i][j] - gamma*gridon[i][j-1]
                gridon[i][j] = gamma*gridon[i][j-1]
                diff+=old
            if grid[i][j]=='U':
                old = gridon[i][j] - gamma*gridon[i-1][j]
                gridon[i][j] = gamma*gridon[i-1][j]
                diff+=old
            else:
                continue
    for i in gridon:
        print(i)
    print(diff)
    if diff==0:
        break


```

```python
dict = [[' ',' ',' ',' '],
        [' ',' ',' ',' '],
        [' ',' ',' ',' ']]
for i in range(0,3):
    for j in range(0,4):
        max = -1
        action = ''
        if i+1 <3 :
            if max<gridon[i+1][j]:
                max = gridon[i+1][j]
                action = "down"
        if i-1>=0:
            if max<gridon[i-1][j]:
                max = gridon[i-1][j]
                action = "up"
        if j-1>=0:
            if max<gridon[i][j-1]:
                max = gridon[i][j-1]
                action = "left"
        if j+1<4:
            if max<gridon[i][j-1]:
                max = gridon[i][j-1]
                action = "right"
        dict[i][j]= action



```
